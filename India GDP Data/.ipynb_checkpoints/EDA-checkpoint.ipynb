```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('India_GDP_Data.csv')
df.head()
```

```python
df.info()
```

```python
df.columns
```

### Lets look at the plots between all variables

```python
sns.pairplot(df)
plt.tight_layout()
```

We can observe that these variables have a good correlation

1) Year and Per Capita Income<br>
2) Year and GDP in Billion USD<br>
3) GDP and Per Capita Income<br>
4) GDP and Percentage Growth<br>

These relations are very evident and conspicous

```python
df.columns
```

```python
sns.lineplot(x='Year',y='GDP_In_Billion_USD',data=df)
```

```python
sns.lineplot(x='Year',y='Per_Capita_in_USD',data=df)
plt.tight_layout()
```

Lets understand the relationship in detail by making a scatterplot

```python
sns.lmplot(x='Year',y='Per_Capita_in_USD',data=df)
```

Seems that Indian economy is making steady progress :)<br>
Now lets look at other relations

```python
df.corr()
```

Lets understand the relationship between different variables with heatmaps

```python
sns.heatmap(df.corr(),annot=True,cmap='rainbow',linewidths=0.5)
```

```python
sns.lmplot(x='GDP_In_Billion_USD',y='Per_Capita_in_USD',data=df)
```

Turns out that GDP and Per Capita Income have a very linear fit and strong correlation (i.e. 1), which implies that Per Capita Income follows the trend of GDP.<br>

So when GDP increases Per Capita Income increases and vice versa.

```python
sns.histplot(x='Year',y='GDP_In_Billion_USD',data=df,bins=10)
plt.tight_layout()
```

Thus we have found the different correlation between the variables.
