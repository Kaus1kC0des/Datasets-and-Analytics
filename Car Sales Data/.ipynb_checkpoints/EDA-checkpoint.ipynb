### Lets have a look at the count of car brands sold

```python
sns.countplot(x='Car Make',data=df)
```

```python
df.groupby('Car Make').count()
```

```python
df.groupby(['Salesperson']).count().sort_values(ascending=False,by='Car Make').head(10)
```

```python
print("From the above dataframe we can find the top ten Sales Executives, wit the highest number of sales being 1229 and the least being 694")
```

```python
print("Lets find the top selling models in each brand")
```

```python
df.groupby(['Car Make',"Car Model"]).max()
```

```python
print("Lets now have a look at the correlation between all the variables")
```

```python
sns.heatmap(df.corr(numeric_only=True),annot=True)
```
